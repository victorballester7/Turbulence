```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 1
t = 100000
nu_1 = 1
nu_2 = 0.01
nu_3 = 0.000001
N = 1000
A = -2
B = 2
```

```python
x = np.linspace(A, B, N)
u_1 = (x - L*np.tanh(L * x / (2 * nu_1 * t))) / t
u_2 = (x - L*np.tanh(L * x / (2 * nu_2 * t))) / t
u_3 = (x - L*np.tanh(L * x / (2 * nu_3 * t))) / t

fig, ax = plt.subplots()

ax.plot(x, u_1, label=r'$\nu = 1$')
ax.plot(x, u_2, label=r'$\nu = 0.01$')
ax.plot(x, u_3, label=r'$\nu = 0.000001$')

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$u^\nu$')
ax.legend()

plt.show()
```

```python
# save figure as .pdf
fig.savefig('burger_t='+str(t)+'.pdf', bbox_inches='tight')
```
